# Collect Data in WebSite

```python
import requests
```

```python
from bs4 import BeautifulSoup
```

```python
r = requests.get('https://www.yelp.com/search?find_desc=tesla+Dealership&find_loc=San+Francisco%2C+CA')
```

```python
r.status_code
```

```python
r.text
```

```python
scoup = BeautifulSoup(r.text,'html.parser')
```

```python
divs = scoup.findAll(class_="css-16lklrv")
```

```python
reviews = []
for div in divs:
    
    reviews.append(div.text)
```

```python
i=0
for review in reviews:
    index = review.find("")
    text=review
    # Supprimer la premire occurrence du mot "cara"
    if index != -1:
        reviews[i] = text[:index] + text[index+len(""):]
    i=i+1

print(reviews)
```

# Sentiment Analysing the Data

```python

import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
import nltk
```

```python
df = pd.DataFrame(np.array(reviews), columns=['reviews'])
```

```python
df['word_count'] = df['reviews'].apply(lambda x: len(x.split()))
```

```python
df['reviews'][0][0]
```

```python
df['char_count'] = df['reviews'].apply(lambda x: len(x))
```

```python
df.head()
```

```python
def average_words(x):
    words = x.split()
    return sum(len(word) for word in words) /len(words) 
```

```python
df['average_words_lenghts'] = df['reviews'].apply(lambda x: average_words(x))
```

```python
df.head()
```

```python
from nltk.corpus import stopwords
```

```python
stop_words = stopwords.words('english')
```

```python
len(stop_words)
```

```python
df['stopword_count'] = df['reviews'].apply(lambda x: len([word for word in x.split() if word.lower() is stop_words]))
```

```python
def stopword_count(df):
    df['stopword_count'] = 0

    for index, row in df.iterrows():
        j = 0
        if isinstance(row['reviews'], str):  # Vrification pour exclure les valeurs None ou NaN
            for word in row['reviews'].split():
                if word.lower() in stop_words:
                    j += 1
        df.at[index, 'stopword_count'] = j

# Utilisez la fonction pour calculer le compte des stopwords
stopword_count(df)

df
```

```python
df['stopword_rate'] = df['stopword_count']/df['word_count']
```

```python
df
```

```python
df.sort_values(by='stopword_rate')
```

```python
df.describe()
```

# Data Cleaning

```python
df['lowercase']=df['reviews'].apply(lambda x:" ".join(word.lower() for word in x.split()))
```

```python
df.head()
```

```python
df['punctuation']=df['lowercase'].str.replace('[^\w\s]','', regex=True)
```

```python
df['stopwords'] = df['punctuation'].apply(lambda x:" ".join(word for word in x.split() if word not in stop_words))
```

```python
pd.Series(" ".join(df['stopwords']).split()).value_counts()[:30]
```

```python
other_stop_words = ['3','would','2015','get','told','week','us','right','left','one','even','also','go','asked']
```

```python
df['cleanreview']=df['stopwords'].apply(lambda x:" ".join(word for word in x.split() if word not in other_stop_words))
```

```python
df.head()
```

# Lemmatization

```python
#import textblob
from textblob import Word
```

```python
df['lemmatized']=df['cleanreview'].apply(lambda x :" ".join(Word(word).lemmatize() for word in x.split()))
```

# Sentiment Analysis

```python
from textblob import TextBlob
```

- `df['polarity'] = df['lemmatized'].apply(lambda x: TextBlob(x).sentiment[0])`: Cette ligne de code ajoute une nouvelle colonne 'polarity' au DataFrame df. Pour chaque valeur dans la colonne 'lemmatized', elle cre un objet TextBlob pour analyser le sentiment et rcupre la valeur de polarit (un nombre compris entre -1 et 1)  l'aide de la mthode .sentiment[0]. La polarit mesure le degr de positivit ou de ngativit du texte.

- `df['subjectivity'] = df['lemmatized'].apply(lambda x: TextBlob(x).sentiment[1])`: Cette ligne de code ajoute une autre nouvelle colonne 'subjectivity' au DataFrame df. Elle fonctionne de manire similaire  la premire ligne, mais cette fois elle rcupre la valeur de subjectivit (un nombre compris entre 0 et 1)  l'aide de la mthode .sentiment[1]. La subjectivit mesure le degr de subjectivit du texte, c'est--dire dans quelle mesure le texte est bas sur des opinions ou des motions plutt que sur des faits objectifs.

```python
df['plarity'] = df['lemmatized'].apply(lambda x:TextBlob(x).sentiment[0])
df['subjectivity'] = df['lemmatized'].apply(lambda x:TextBlob(x).sentiment[1])
```

```python
df.drop(['lowercase','punctuation','stopwords','cleanreview','lemmatized'],axis=1,inplace=True)
```
